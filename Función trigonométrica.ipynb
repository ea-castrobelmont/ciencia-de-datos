```python
#importando librerías
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definición de la función sinusoidal
def temp_diaria_promedio(t):
    A = (30 - 15)/2
    B = 15 + A
    w = 2*np.pi/24
    return A*np.sin(w*t) + B
```

```python
# Creación de los datos de temperatura diaria promedio
t = np.linspace(0, 365, 366)
temp = temp_diaria_promedio(t)
```

```python
# Visualización de los datos
plt.plot(t, temp)
plt.title('Temperatura diaria promedio en Ciudad X')
plt.xlabel('Días')
plt.ylabel('Temperatura (°C)')
plt.show()
```
